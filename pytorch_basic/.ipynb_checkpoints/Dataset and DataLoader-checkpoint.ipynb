```python
import numpy as np
import torch
from torch.utils.data import Dataset #抽象类只能继承
from torch.utils.data import DataLoader #可以实例化
from tqdm import tqdm
import matplotlib
import matplotlib.pyplot as plt
```

```python
#加载数据部分
class DiabetesDataset(Dataset):
    def __init__(self,filepath):
        xy=np.loadtxt(filepath,delimiter=',',dtype=np.float32) #读取数据转化为numpy数据
        self.len=xy.shape[0] #取第一个元素也就是数据数量
        self.x_data=torch.from_numpy(xy[:,:-1]) #读取x的数据，是取所有的行，第一列到倒数第二列
        self.y_data=torch.from_numpy(xy[:,[-1]]) #读取y的数据，是取所有的行，倒数第一列
    def __getitem__(self,index):
       return self.x_data[index],self.y_data[index]
    def __len__(self):
        return self.len
dataset=DiabetesDataset('diabetes.csv.gz')
train_data=DataLoader(dataset=dataset,batch_size=32,shuffle=True,num_workers=0)
```

```python
#设计模型部分
class Model(torch.nn.Module):
    def __init__(self):
        super(Model,self).__init__()
        self.linear1=torch.nn.Linear(8,6)
        self.linear2=torch.nn.Linear(6,4)
        self.linear3=torch.nn.Linear(4,1)
        self.sigmoid=torch.nn.Sigmoid()
    def forward(self,x):
        x=self.sigmoid(self.linear1(x))
        x=self.sigmoid(self.linear2(x))
        x=self.sigmoid(self.linear3(x))
        return x
model=Model()
```

```python
#设计损失和优化
criterion=torch.nn.BCELoss(size_average=True)
optimizer=torch.optim.SGD(model.parameters(),lr=0.001)
```

```python
#训练过程
if __name__ == '__main__':
    for epoch in tqdm(range(100)):
        for i,data in enumerate(train_data,0):
        #Prepare Data
            input,labels=data
        #Forward
            y_pred=model(input)
            loss=criterion(y_pred,labels)
            print(epoch,i,loss.item())
        #Backward
            optimizer.zero_grad()
            loss.backward()
        #Updata
            optimizer.step()
```
