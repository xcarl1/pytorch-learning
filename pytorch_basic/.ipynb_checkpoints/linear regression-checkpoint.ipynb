```python
from matplotlib import pyplot as plt
import torch
import random
```

```python
def synthetic_data(w,b,num_examples): #根据带有噪声的线性模型构造一个人造数据集，w=[2,-3.4],b=4.2
    """y=Xw+b+噪声"""
    X=torch.normal(0,1,(num_examples,len(w))) #生成均值为0，方差为1，形状为样本数*w行数的矩阵，作为样本数据的x
    y= torch.matmul(X , w) + b #计算预测值y
    y+=torch.normal(0,0.01,y.shape) #加入噪声
    return X,y.reshape((-1,1)) #reshape((-1,1))表示将矩阵转化成列向量
true_w=torch.tensor([2,-3.4])
true_b=4.2
features,labels=synthetic_data(true_w,true_b,1000)
```

```python
print("features:",features[0],"labels:",labels[0]) #看一下数据
```

```python
plt.scatter(features[:,1].detach().numpy(),labels.detach().numpy(),20) #X的第二个元素与y之间的散点图
```

```python
"""定义一个data_iter函数，接受批量大小，特征矩阵，标签向量作为输入，生成大小为batch_size大小的批量"""
def data_iter(batch_size,features,labels):
    num_examples=len(features) #样本数量
    indices=list(range(num_examples)) #num_examples个标签
    random.shuffle(indices)
    for i in range(0,num_examples,batch_size): #从0到num_examples,每次跳batch_size个单位
        batch_indices=torch.tensor(indices[i:min(i+batch_size,num_examples)])
        yield features[batch_indices],labels[batch_indices]
```

```python
batch_size=10
for X,y in data_iter(batch_size,features,labels):
    print(X,"\n",y)
    break
```

```python
"""训练数据定义好之后，下面把模型定义好"""
w=torch.normal(0,0.01,size=(2,1),requires_grad=True)
b=torch.zeros(1,requires_grad=True)
```

```python
"""定义模型"""
def linreg(X,w,b):
    """线性回归模型"""
    return torch.matmul(X,w)+b #matmul函数是矩阵乘法的意思
```

```python
"""定义损失函数"""
def squared_loss(y_hat,y): #y_hat是预测值，y是真实值
    """均方损失"""
    return (y_hat-y.reshape(y_hat.shape))**2/2 #没有除n
```

```python
"""定义优化算法"""
def sgd(params,lr,batch_size): #params是一个链表，包括w和b；lr是学习率；batch_size是批量大小
    """小批量随机梯度下降"""
    with torch.no_grad():
        for param in params:
            param-=lr*param.grad/batch_size
            param.grad.zero_()
```

```python
"""训练过程"""
lr=0.04
epochs_num=3 #进行次数
net=linreg #模型
loss=squared_loss

for epoch in range(epochs_num):
    for X,y in data_iter(batch_size,features,labels):
        l=loss(net(X,w,b),y) #X和y的小批量损失
        l.sum().backward()
        sgd([w,b],lr,batch_size) #使用参数的梯度更新参数
    with torch.no_grad():
        train_1=loss(net(features,w,b),labels)
        print(f'epoch{epoch+1},loss{train_1.mean():f}')
```
