```python
import torch
import math
import numpy as np
from d2l import torch as d2l
from torch import nn
```

```python
#生成人工数据
max_degree=20
num_train=100
num_test=100
true_w=np.zeros(max_degree)
true_w[0:4]=np.array([5,1.2,-3.4,5.6])
true_w
```

```python
features=np.random.normal(size=(num_train+num_test,1))
np.random.shuffle(features)
poly_features=np.power(features,np.arange(max_degree).reshape(1,-1))
poly_features.shape
```

```python
for i in range(max_degree):
    poly_features[:,i]/=math.gamma(i+1)
labels=np.dot(poly_features,true_w)
labels+=np.random.normal(scale=0.1,size=labels.shape)
true_w, features, poly_features, labels = [torch.tensor(x,dtype=torch.float32) for x in [true_w, features, poly_features, labels]]
```

```python
#定义一个评估函数
def evaluate_loss(net,data_iter,loss):
    metric=d2l.Accumulator(2)
    for X,y in data_iter:
        X=X.to(torch.float32)
        y_hat=net(X)
        y=y.reshape(y_hat.shape)
        l=loss(y_hat,y)
    return metric[0]/metric[1]
```

```python
#训练过程
def train(train_features,test_features,train_labels,test_labels,epochs=400):
    loss=nn.MSELoss()
    input_shape=train_features.shape[-1]
    batch_sizes=min(10,train_features.shape[0])
    net=nn.Linear(in_features=input_shape,out_features=1,bias=False)
    train_iter=d2l.load_array((train_features,train_labels.reshape(-1,1)),batch_sizes)
    test_iter=d2l.load_array((test_features,test_labels.reshape(-1,1)),batch_sizes,is_train=False)
    animator = d2l.Animator(xlabel='epoch',ylabel='loss',yscale='log',xlim=[1,epochs],ylim=[1e-4,1e2],legend=['train','test'])
    trainer=torch.optim.SGD(net.parameters(),lr=0.01)
    for epoch in range(epochs):
        d2l.train_epoch_ch3(net,train_iter,loss,trainer)
        if epoch==0 or (epoch+1)%20==0:
            animator.add(epoch+1,(evaluate_loss(net,train_iter,loss),evaluate_loss(net,test_iter,loss)))
    print('weight',net.weight.data.numpy())
```

```python
#三阶多项式拟合
train(poly_features[:num_train,:4],poly_features[num_train:,:4],labels[:num_train],labels[num_train:])
```

```python
#一阶多项式拟合
train(poly_features[:num_train,:2],poly_features[num_train:,:2],labels[:num_train],labels[num_train:])
```

```python
#一阶多项式拟合
train(poly_features[:num_train,:],poly_features[num_train:,:],labels[:num_train],labels[num_train:])
```
